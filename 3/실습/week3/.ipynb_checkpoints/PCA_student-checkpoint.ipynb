```python
# Hint
# plt.legend(), plt.title(), plt.scatter()  

def show2D(pca_score, Y, target_names):
    plt.figure()
    # TODO -->
    tmp = PCA(n_components=2)
    pca_1 = tmp.fit(pca_score).transform(pca_score)
    for i, target_name in zip(range(len(target_names)), target_names):
        plt.scatter(pca_1[Y == i, 0], pca_1[Y == i, 1], label=target_name)
    plt.legend(loc = 'lower right')
    plt.title('Results', fontsize = 12)
    plt.show()
    # <--
```

```python
show2D(PC_score, Y, target_names)
```

```python
def student_pca(X, n_components=2):
    pca_results = None
    # TODO -->
    iris_cov = np.cov(X.T)
    eig_vals, eig_vecs = np.linalg.eig(iris_cov)
    eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]
    eig_pairs[::-1]
    matrix_w = np.hstack((eig_pairs[0][1].reshape(4,1), eig_pairs[1][1].reshape(4,1)))
    Y = X.dot(matrix_w)
    pca_results = Y
    # <--
    return pca_results
    
pca_student_score = student_pca(X, n_components=2)
```

```python
show2D(pca_student_score, Y, target_names)
```
